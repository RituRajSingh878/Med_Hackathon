```python
new_data = {}
for i in sorted_items:
    if i[1]>1500:
        new_data[i[0]]=[]
```

```python
ct = 1
nm = 10;
le = 0 
for enc in obs['ENCOUNTER'].unique():
    data = obs[obs['ENCOUNTER']==enc]
    if ct%nm==0:
        print(ct)
        nm = nm*10
    for i in range(len(data)):
        if new_data.get(data.iloc[i]['DESCRIPTION']) != None:
            new_data[data.iloc[i]['DESCRIPTION']].append([data.iloc[i]['VALUE']][0])
    for d, _ in new_data.items():
        if len(_)<le:
            new_data[d].append(np.nan)
    le = le+1
    ct = ct+1
```

```python
from sklearn import preprocessing
import numpy as np
scaled = preprocessing.scale(train)
scaled
```

# Clustering Algo

```python
from sklearn.cluster import KMeans

distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k, random_state=0)
    kmeanModel.fit(scaled)
    distortions.append(kmeanModel.inertia_)

import matplotlib.pyplot as plt
%matplotlib inline 

plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()    

```

```python
#bset with 2 clusters
kmodel = KMeans(n_clusters=3, random_state=42)
kmodel.fit(scaled)

from pickle import dump
dump(kmodel, open('kmodel.pkl', 'wb'))
```

```python
ct1 = 0
ct2 = 0
ct3 = 0
for i in kmodel.predict(scaled):
    if i == 0:
        ct1 = ct1+1
    elif i == 1:
        ct2 = ct2+1
    elif i == 2:
        ct3 = ct3+1
print(ct1, ct2, ct3)        
```
